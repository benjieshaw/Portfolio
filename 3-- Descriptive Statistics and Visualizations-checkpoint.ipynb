# Visualizations

Let's start with histograms of the numerical values.

```python
#Create subplots
fig, axs = plt.subplots(3, 2, figsize=(12,10))

#Flatten the axs for iteration
axs = axs.flatten()

#List the columns to plot
cols = ['Weight', 'NV', 'Visc-Ford', 'Max_Temp_Degrees', "Rain_Inches"]

#Plot histograms
for i, col in enumerate(cols):
    full_data[col].plot(kind='hist', ax=axs[i])
    
    #Calculate and plot the mean for each histogram
    mean_value = full_data[col].mean()
    axs[i].axvline(mean_value, color='r', linestyle='dashed', linewidth=2)
    axs[i].text(mean_value, axs[i].get_ylim()[1]*0.9, f'Mean: {mean_value:2f}', color='r')
    
    axs[i].set_title(f'Hisogram of {col}')
    axs[i].set_xlabel(f'Values of {col}')
    axs[i].set_ylabel('Frequency')

#Adjust the layout to prevent overlap
plt.tight_layout()

#Hide unused plots
for j in range(i+1, len(axs)):
    fig.delaxes(axs[j])

plt.show()
```

### Observations
•As expected from the descriptive statistics, the production data provided by the QA manager at the plant doesn't evidence a wide range of values in data. The weight histogram is slightly skewed to the left. It's difficult to visually identify skew in the NV histogram because the values are so condensed.

•For temperature data, max temp is heavily skewed to the left. This makes sense given that the facility is in Ohio where the weather tends to be more cold. 

•Rain data is slightly skewed to the right. This might suggest that rain tends to be more of an all-or-nothing proposition in this area of hte country. Either it rains a good deal, or there is no rain.

## Time Series

It would be helpful to see how often the viscosity values fall outside of acceptable viscocity ranges. Since the data covers a wide range of dates, we will utilize a time series analysis to visualize when viscosity falls out of the control and/or specification limits.

Also, some of the dates covered in the data are from a long time ago. For the purposes of this visualization, we will restrict the date range of the data to the previous 6 years from the latest date (3/8/24).

```python
# Convert 'Date' column to datetime
full_data['Date'] = pd.to_datetime(full_data['Date'], format='%m/%d/%y')

# Define upper and lower value ranges
UCL = 18.3
USL = 19
LCL = 16.3
LSL = 15

# Create a new column to identify out-of-range values
full_data['Out_of_Range'] = (full_data['Visc-Ford'] < LCL) | (full_data['Visc-Ford'] > UCL)

# Filter the dates
start_date = '2018-03-08'
end_date = '2024-03-08'
vis_data = full_data[(full_data['Date'] >= start_date) & (full_data['Date'] <= end_date)] 

# Set the Date column as the index
vis_data.set_index('Date', inplace=True)

# Plotting
plt.figure(figsize=(20, 10))

# Plot all data points
sns.lineplot(data=vis_data, x=vis_data.index, y='Visc-Ford', label='Viscosity', color='lightgrey')

# Highlight out-of-range points
sns.scatterplot(data=vis_data[vis_data['Out_of_Range']], x=vis_data[vis_data['Out_of_Range']].index, y='Visc-Ford', color='red', label='Out of Range')

# Plot the acceptable range limits
plt.axhline(LCL, color='green', linestyle='--', label='Lower Control Limit')
plt.axhline(UCL, color='green', linestyle='--', label='Upper Control Limit')
plt.axhline(USL, color='maroon', linestyle='dashdot', label='Upper Spec Limit')
plt.axhline(LSL, color='maroon', linestyle='dashdot', label='Lower Spec Limit')


# Formatting the x-axis to show dates
date_format = mdates.DateFormatter('%m-%d-%y')
plt.gca().xaxis.set_major_formatter(date_format)

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Viscosity')
plt.title('Time Series of Viscosity with Out-of-Range Points Highlighted')
plt.legend(loc='upper right')

# Show plot
plt.tight_layout()
plt.show()
```

### Analysis

As a reminder, the Specification Limits are limits that are acceptable to the client. The Control Limits are acceptable to the plant as a production standard. 

Over this 6-year period, it was rare for the chemical to fall below or rise above the specification standards. However, falling outside of the control standards, especially falling below the Control Limit, occurred far more often.
